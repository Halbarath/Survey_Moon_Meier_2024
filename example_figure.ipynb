```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fs_label = 16
fs = 13
fs_legend = 12
fw = 'normal'
cmap_standard = plt.cm.turbo
cmap_secondary = plt.cm.viridis

figwidth = 13

axis_labels = {'planet_disk_mixing_delta': r"$\delta_{pd}$",
               'M_disk': r"$M_{d}$ $(M_{☾})$",
               'Fe_disk': r"$F_{d}^{Fe}$",
               'J_bound': r"$J_{b}$ $(J_{EM})$",
               'b_inf': r"$b_{\infty}$ $(R_{grav})$",
               'v_inf': r"$v_{\infty}$ $(v_{esc})$",
               'dir_J_tot_init': r"$J_{0}$ $(J_{EM})$",
               'accretion_efficiency': r"$\eta$",
               'P_rot': r"$P_{rot}$ $(hrs)$",
               'dir_gamma': r"$\gamma$",
               'planet_disk_mixing_delta_abs': r"$\left|\delta_{pd}\right|$",
               'J_tot_init': r"$J_{0}$ $(J_{EM})$",
               'dir_omega': r"$f_{\Omega}$",
               'HSSL': r'$R_P/R_{HSSL}$',
}

markers = ["o","v","^","<",">","1","2","3","4","s","p"]
```

```python
# Current Earth-Moon angular momentum [kg m^2 s^-1]
J_EM = 3.61e34
```

```python
data = pd.read_csv('collision_analysis.csv')
data['J_bound'] = data['J_bound']/J_EM
data['J_tot_init'] = data['J_tot_init']/J_EM
data['Fe_disk'] = np.max([data['Fe_disk'],np.ones_like(data['Fe_disk'])*1e-4],0)
data.drop('omega_sb', inplace=True, axis=1) # column is unused
data.drop('P_rot_sb', inplace=True, axis=1) # column is unused
fulldata = data
data
```

```python
df = data.copy()
df.sort_values(by=['v_inf'],inplace=True,ascending=True)
df = df[(df['M_bound'] >= 1.0) & (df['M_disk'] >= 0.0)] # Select Mergers

print('Number of cases: {}'.format(len(df)))
figsize = (figwidth,figwidth*1.0)
fig1 = plt.figure(figsize=figsize)
axs = fig1.subplots(3,3,sharex='col',sharey='row')

k,l = 0,0
Xvalue = 'J_bound'
Yvalue = 'M_disk'
for i,gamma in enumerate(np.sort(np.unique(fulldata['dir_gamma']))[::-1]):
    if gamma in df['dir_gamma'].unique():
        df1 = df[df['dir_gamma'] == gamma]
        axs[k,l].scatter(df1[Xvalue],df1[Yvalue], marker=markers[i],c=df1['v_inf'], vmin=0, vmax=3,cmap=cmap_standard)
axs[k,l].set_ylabel(axis_labels[Yvalue], fontsize=fs_label)
axs[k,l].tick_params(bottom=False,labelbottom=False,left=True,labelleft=True)
axs[k,l].set_xlim(left=-0.1, right=3.4)
axs[k,l].set_ylim(bottom=-0.1, top=7.1)

k,l = 1,0
Xvalue = 'J_bound'
Yvalue = 'Fe_disk'
for i,gamma in enumerate(np.sort(np.unique(fulldata['dir_gamma']))[::-1]):
    if gamma in df['dir_gamma'].unique():
        df1 = df[df['dir_gamma'] == gamma]
        axs[k,l].scatter(df1[Xvalue],df1[Yvalue], marker=markers[i],c=df1['v_inf'], vmin=0, vmax=3,cmap=cmap_standard)
axs[k,l].set_ylabel(axis_labels[Yvalue], fontsize=fs_label)
axs[k,l].set_yscale('log')
axs[k,l].tick_params(bottom=False,labelbottom=False,left=True,labelleft=True)
axs[k,l].set_xlim(left=-0.1, right=3.4)
axs[k,l].set_ylim(bottom=1e-4,top=2e0)

k,l = 2,0
Xvalue = 'J_bound'
Yvalue = 'planet_disk_mixing_delta'
for i,gamma in enumerate(np.sort(np.unique(fulldata['dir_gamma']))[::-1]):
    if gamma in df['dir_gamma'].unique():
        df1 = df[df['dir_gamma'] == gamma]
        axs[k,l].scatter(df1[Xvalue],df1[Yvalue], marker=markers[i],c=df1['v_inf'], vmin=0, vmax=3,cmap=cmap_standard)
axs[k,l].set_xlabel(axis_labels[Xvalue], fontsize=fs_label)
axs[k,l].set_ylabel(axis_labels[Yvalue], fontsize=fs_label)
axs[k,l].tick_params(bottom=True,labelbottom=True,left=True,labelleft=True)
axs[k,l].set_xlim(left=-0.1, right=3.4)
axs[k,l].set_ylim(bottom=-0.65, top=1.05)

Xvalue = 'M_disk'
Yvalue = 'Fe_disk'    
k,l = 1,1
for i,gamma in enumerate(np.sort(np.unique(fulldata['dir_gamma']))[::-1]):
    if gamma in df['dir_gamma'].unique():
        df1 = df[df['dir_gamma'] == gamma]
        axs[k,l].scatter(df1[Xvalue],df1[Yvalue], marker=markers[i],c=df1['v_inf'], vmin=0, vmax=3,cmap=cmap_standard)
axs[k,l].set_xlabel(axis_labels[Xvalue], fontsize=fs_label)
axs[k,l].set_yscale('log')
axs[k,l].tick_params(bottom=False,labelbottom=False,left=False,labelleft=False)
axs[k,l].tick_params(which='minor',left=False)
axs[k,l].set_xlim(left=-0.1, right=7.1)
axs[k,l].set_ylim(bottom=0.5e-4,top=2e0)

k,l = 2,1
Xvalue = 'M_disk'
Yvalue = 'planet_disk_mixing_delta'
for i,gamma in enumerate(np.sort(np.unique(fulldata['dir_gamma']))[::-1]):
    if gamma in df['dir_gamma'].unique():
        df1 = df[df['dir_gamma'] == gamma]
        axs[k,l].scatter(df1[Xvalue],df1[Yvalue], marker=markers[i],c=df1['v_inf'], vmin=0, vmax=3,cmap=cmap_standard)
axs[k,l].set_xlabel(axis_labels[Xvalue], fontsize=fs_label)
axs[k,l].tick_params(bottom=True,labelbottom=True,left=False,labelleft=False)
axs[k,l].set_xlim(left=-0.1, right=7.1)
axs[k,l].set_ylim(bottom=-0.65, top=1.05)

k,l = 2,2
Xvalue = 'Fe_disk'
Yvalue = 'planet_disk_mixing_delta'
for i,gamma in enumerate(np.sort(np.unique(fulldata['dir_gamma']))[::-1]):
    if gamma in df['dir_gamma'].unique():
        df1 = df[df['dir_gamma'] == gamma]
        axs[k,l].scatter(df1[Xvalue],df1[Yvalue], marker=markers[i],c=df1['v_inf'], vmin=0, vmax=3,cmap=cmap_standard)
axs[k,l].set_xlabel(axis_labels[Xvalue], fontsize=fs_label)
axs[k,l].set_xscale('log')
axs[k,l].tick_params(bottom=True,labelbottom=True,left=False,labelleft=False)
axs[k,l].set_xlim(left=0.5e-4, right=2e0)
axs[k,l].set_ylim(bottom=-0.65,top=1.05)

fig1.delaxes(axs[0][1])
fig1.delaxes(axs[0][2])
fig1.delaxes(axs[1][2])

fig1.subplots_adjust(wspace=0.0,hspace=0.0)

PCM=axs[2,2].get_children()[0]
cax = axs[2][2].inset_axes([0.7, 1.05, 0.1, 1.9])
cbar = plt.colorbar(mappable=PCM,cax=cax)
cbar.ax.tick_params(labelsize=fs)
cbar.ax.set_ylabel(axis_labels['v_inf'], fontsize=fs_label)

for ax in axs.reshape(-1):
    for tick in ax.xaxis.get_major_ticks():
        tick.label1.set_fontsize(fs)
        tick.label1.set_fontweight(fw)
    for tick in ax.yaxis.get_major_ticks():
        tick.label1.set_fontsize(fs)
        tick.label1.set_fontweight(fw)

plt.axes(axs[0,0])
for i,gamma in enumerate(np.sort(np.unique(fulldata['dir_gamma']))[::-1]):
    if gamma in df['dir_gamma'].unique():
        plt.scatter([],[],color='k',marker=markers[i],label=gamma)
    plt.legend(title='$\gamma$',prop={'size': fs_legend}, title_fontsize=fs_legend,ncol=2,bbox_to_anchor=[1.1,1.0],frameon=False)

fig1.savefig("example.png", format='png', bbox_inches='tight', dpi=300)
```
